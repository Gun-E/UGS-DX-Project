# Bing 크롤링
## 키워드 : 광파오븐

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Chrome WebDriver 설정
driver = webdriver.Chrome()

# 빈 리스트 생성 (크롤링한 데이터를 저장하기 위해)
data = []

# 페이지 번호 1부터 100까지 반복
for page in range(1, 10001, 10):
    url = f"https://www.bing.com/search?q=%ea%b4%91%ed%8c%8c%ec%98%a4%eb%b8%90+%ed%9b%84%ea%b8%b0&qs=n&sp=-1&lq=0&pq=%ea%b4%91%ed%8c%8c%ec%98%a4%eb%b8%90+%ed%9b%84%ea%b8%b0&sc=10-7&sk=&cvid=5A22F67EB53444CDB54190306EFF25D5&ghsh=0&ghacc=0&ghpl=&FPIG=EE8CE2EC693F4437ACC6CDD2B7DE3A88&FPIG=4458245AD3CD4A8FA3261639F8F07897&first={page}"
    # 페이지 이동
    driver.get(url)
    driver.maximize_window()
    
    try:
        time.sleep(2)
        articles = driver.find_elements(By.CSS_SELECTOR, "#b_results > li > div > p")
        for article in articles:
            title = article.text
            print(title)
            data.append(title)  # 제목을 리스트에 저장
            
    except Exception as e:
        print(f"Error on page {page}: {e}")
        
        # 에러가 발생하면 수집된 데이터를 CSV 파일로 저장
        df = pd.DataFrame(data, columns=["Title"])
        df.to_csv(f"bing_{page}.csv", index=False)
        print(f"Data saved until page {page}. Exiting.")
        break  # 오류 발생 시 크롤링 종료
    
    # 페이지 로딩 사이에 잠시 대기
    time.sleep(2)

# 마지막으로 수집된 데이터를 저장
if len(data) > 0:
    df = pd.DataFrame(data, columns=["Title"])
    df.to_csv("data/bing.csv", index=False)
    print("Final data saved.")

# 브라우저 종료
driver.quit()
```
