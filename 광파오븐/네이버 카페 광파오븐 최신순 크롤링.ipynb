```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Chrome WebDriver 설정
driver = webdriver.Chrome()

# 빈 리스트 생성 (크롤링한 데이터를 저장하기 위해)
data = []

# 페이지 번호 1부터 100까지 반복
for page in range(1, 101):
    url = f"https://section.cafe.naver.com/ca-fe/home/search/articles?q=%EA%B4%91%ED%8C%8C%EC%98%A4%EB%B8%90&t=1728984004589&em=1&od=1&p={page}"
    
    # 페이지 이동
    driver.get(url)
    driver.maximize_window()
    
    try:
        time.sleep(2)
        articles = driver.find_elements(By.CSS_SELECTOR, "#mainContainer > div.content > div.section_home_search > div.search_item_wrap > div.item_list > div > div > div > a:nth-child(1) > p")
        
        for article in articles:
            title = article.text
            print(title)
            data.append(title)  # 제목을 리스트에 저장
            
    except Exception as e:
        print(f"Error on page {page}: {e}")
        
        # 에러가 발생하면 수집된 데이터를 CSV 파일로 저장
        df = pd.DataFrame(data, columns=["Title"])
        df.to_csv(f"naver_cafe_articles_until_page_{page}.csv", index=False)
        print(f"Data saved until page {page}. Exiting.")
        break  # 오류 발생 시 크롤링 종료
    
    # 페이지 로딩 사이에 잠시 대기
    time.sleep(2)

# 마지막으로 수집된 데이터를 저장
if len(data) > 0:
    df = pd.DataFrame(data, columns=["Title"])
    df.to_csv("naver_cafe_articles_final_new.csv", index=False)
    print("Final data saved.")

# 브라우저 종료
driver.quit()
```
